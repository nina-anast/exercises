## Άσκηση 2.2

````python
message = input("Εισάγετε το μήνυμα σας: ").lower()
vowels = "αεηιουωάέήίόύώϊϋ"
num_vowels = 0
void = 0
for char in message:
    if char == " ":
        void +=1
    if char in vowels:
        num_vowels +=1
pososto = num_vowels / (len(message)- void)
print("Ο αριθμός των φωνηέντων εντός του μηνύματος ισούται με: ",
num_vowels,'\n')
print("Αποτελεί το", pososto)
print(len(message))
print(void)
````

**Διαμορφώστε τον κώδικα έτσι ώστε να υπολογίζει και τον αριθμό των σημείων στίξεως που
θα βρεθούν στο κείμενο που δώσατε.**

Οι μετατροπές που θα πραγματοποιήσουμε θα είναι για να ακολουθήσουμε την ίδια διαδικασία με τα φωνήεντα και για τα σημεία στίξης. Δηλαδή προσθέσαμε μια μεταβλητή συμβολοσειράς με όλα τα σημεία στίξης (```marks```). Στην συνέχεια, αρχικοποιήσαμε τον αριθμό των σημείων στίξης της πρότασης να είναι ίσος με μηδέν (```num_marks```). Και στην συνέχεια, ελέγχουμε κάθε χαρακτήρα για να δούμε αν είναι σημείο στίξης και όποτε βρίσκουμε έναν, αυξάνουμε τον αριθμό ```num_marks``` κατά ένα. Τέλος, υπολογίζουμε το ποσοστό της πρότασης που είναι τα σημεία στίξης ```pososto_marks``` και εμφανίζουμε τα αποτελέσματα.

Μια αλλαγή που έγινε ακόμα είναι το elif, για να μην ελέγχει αν ένας χαρακτήρας είναι σημείο στίξης, αν έχει ήδη αναγνωριστεί ως κενό η φωνήεν. Ακόμα, κρατήσαμε λιγότερα σημαντικά ψηφία στα ποσοστά που εμφανίζουμε στον χρήστη μέσω των μεταβλητών ```formatted_vowels``` και ```formatted_marks```.

````python
message = input("Εισάγετε το μήνυμα σας: ").lower()
vowels = "αεηιουωάέήίόύώϊϋ"
marks = ".,;:!?·"     # λίστα σημείων στίξης 
num_vowels = 0
num_marks = 0         # αρχικοποιούμε τον αριθμό των σημείων στίξης του κειμένου
void = 0
for char in message:
    if char == " ":
        void +=1
    elif char in vowels:
        num_vowels +=1 
    elif char in marks:
        num_marks +=1 # ελέγχουμε κάθε χαρακτήρα που δεν είναι κενός ή φωνήεν, 
                      # αν είναι σημείο στίξης
pososto_vowels = num_vowels / (len(message)- void)
formatted_vowels = format(pososto_vowels,".3f")  # μειώνουμε τα δεκαδικά ψηφία του ποσοστού 
                                                 # σε 3 για να είναι πιο καθαρή η έξοδος 
                                                 # για τον χρήστη
pososto_marks = num_marks / (len(message)- void) # βρίσκουμε το ποσοστό των σημείων στίξης
formatted_marks = format(pososto_marks,".3f")   
print("Ο αριθμός των φωνηέντων εντός του μηνύματος ισούται με: ",
      num_vowels)
print("Αποτελεί το", formatted_vowels, "% \n")
print("Αντίστοιχα, ο αριθμός των σημείων στίξης εντός του μηνύματος ισούται με ", 
      num_marks)
print("Αποτελεί το", formatted_marks, "% \n")
print("Το συνολικό μέγεθος του μηνύματος είναι" , len(message))
print("Και ο αριθμός των κενών χαρακτήρων του μηνήματος",void)
````

## Άσκηση 3

Διαμορφώστε εκ νέου μια καινούργια λίστα με 20 τιμές θερμοκρασίας όπου θα υπάρχουν 
τουλάχιστον δυο τιμές -9999. Διαμορφώστε τον κώδικα ώστε να βρίσκεται σε ποια ημέρα του
μήνα καταγράφηκαν οι τιμές -9999 και δημιουργήστε εκ νέου το σχετικό γράφημα διασποράς –
γραμμής. Δίνετε ο κώδικας για το πρώτο μέρος της άσκησης. 

````python
# Δημιουργία λίστας με τιμές θερμοκρασίας
temp = [16, 17, 15, 14, 13, -9999, 15, 14, 17, 18]
# Εύρεση της ημέρας με τη λανθασμένη τιμή
nontempday = temp.index(-9999) + 1 # Προσθέτουμε 1 για να μετράμε από 1
                                   # αντί για 0
print(f"Η λανθασμένη τιμή καταγράφηκε την {nontempday}η Μαρτίου.")
# Αφαίρεση της λανθασμένης τιμής από τη λίστα
cleantemp = [t for t in temp if t != -9999]
# Δημιουργία γραφήματος
import matplotlib.pyplot as plt
days = list(range(1, 11)) # Δημιουργία λίστας για τις ημέρες του Μαρτίου
days.remove(nontempday) # Αφαίρεση της ημέρας με τη λανθασμένη τιμή
plt.figure(figsize=(10, 6))
plt.plot(days, cleantemp, marker='o', linestyle='-', color='blue')
plt.title('Θερμοκρασίες Μαρτίου (χωρίς τη λανθασμένη τιμή)')
plt.xlabel('Ημέρα')
plt.ylabel('Θερμοκρασία (°C)')
plt.grid(True)
plt.xticks(range(1, 11)) # Εμφάνιση όλων των ημερών στον άξονα x
plt.show()
````

**Ο τροποποιημένος κώδικας παρατίθεται παρακάτω.**

Με την βιβλιοθήκη ```random``` δημιουργούμε ένα διάνυσμα με 20 τιμές (```range(20)```) με τις νέες τιμές της θερμοκράσιας, σε ένα εύρος τιμών $15-20^{ο}C$ (```random.randint(15, 22)```). Επιλέγουμε ξανά με την εντολή ```error_days = random.randint(2, 20)``` ένα τυχαίο πλήθος ημερών στο εύρος [2,20] για το οποίο θα λαμβάνουμε λανθασμένες τιμές. Στην συνέχεια, επιλέγουμε τυχαίες θέσεις από το διάνυσμα (```indices_to_replace = random.sample(range(len(temp)), error_days)```) και τοποθετούμε σε αυτές την τιμή -9999. Στην συνέχεια, εμφανίζουμε τις τιμές των θερμοκρασιών που λάβαμε.

Τροποποιούμε την εντολή ```nontempday = temp.index(-9999)``` και τοποθετούμε στην θέση της την ```nontempdays = [i + 1 for i, value in enumerate(temp) if value == -9999]```. Αυτή η εντολή ελέγχει κάθε τιμή του διανύσματος ```temp``` αν είναι -9999. Αν είναι προσθέτει 1 στον δείκτη και αποθηκεύει στην πρώτη θέση την τιμή πριν προχωρήσει στην επόμενη.

Στην συνέχεια, ακολουθούμε τον κώδικα της εκφώνησης και αφαιρούμε τις λανθασμένες τιμές και παρουσιάζουμε το διάγραμμα. Μια σημαντική αλλαγή που έπρεπε να κάνουμε είναι να αλλάξουμε το μέγεθος της λίστας των ημερών και να γίνει ```days = list(range(1, 21))```.

````python
import random

# Πρόσθέτουμε θερμοκρασίες σε ένα αυθαίρετο εύρος 15-22oC
temp = [random.randint(15, 22) for _ in range(20)]

# Ορίζουμε τον αριθμό των ημερών με λανθασμένες τιμές
error_days = random.randint(2, 20)

# Κάνω τιμές σε τυχαίες θέσεις ίσες με -9999 
indices_to_replace = random.sample(range(len(temp)), error_days)
for index in indices_to_replace:
    temp[index] = -9999

# Εμφανίζω τις τιμές
print("Οι θερμοκρασίες που προκύψαν είναι : \n ",temp)

# Βρίσκουμε όλες τις λανθασμένες τιμές
nontempdays = [i + 1 for i, value in enumerate(temp) if value == -9999]

# Εμφανίζουμε τις μέρες που λάβαμε τις λανθασμένες τιμές
for index in nontempdays:
    print(f"Η λανθασμένη τιμή καταγράφηκε την {index}η Μαρτίου.")

# Αφαίρεση της λανθασμένης τιμής από τη λίστα
cleantemp = [t for t in temp if t != -9999]

# Δημιουργία γραφήματος
import matplotlib.pyplot as plt
days = list(range(1, 21)) # Δημιουργία λίστας για τις ημέρες του Μαρτίου
for nontempday in nontempdays:
    days.remove(nontempday) # Αφαίρεση της ημέρας με τη λανθασμένη τιμή
plt.figure(figsize=(10, 6))
plt.plot(days, cleantemp, marker='o', linestyle='-', color='blue')
plt.title('Θερμοκρασίες Μαρτίου (χωρίς τη λανθασμένη τιμή)')
plt.xlabel('Ημέρα')
plt.ylabel('Θερμοκρασία (°C)')
plt.grid(True)
plt.xticks(range(1, 11)) # Εμφάνιση όλων των ημερών στον άξονα x
plt.show()
````
